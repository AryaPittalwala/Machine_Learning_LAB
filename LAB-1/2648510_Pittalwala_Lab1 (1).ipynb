# **Task 6 — Wrap it in a function (Functions)**


```python
def build_xy_matrices(records_list):
    """
    Builds the feature matrix X and target vector y for machine learning.

    X consists of 'Transaction Amount', 'Merchant Category Code (MCC)', and 'Transaction Response Code'.
    y consists of 'Fraud Flag or Label' (binary encoded).

    Args:
        records_list (list): A list of dictionaries, where each dictionary represents a record
                             and contains relevant fields.

    Returns:
        tuple: A tuple containing:
            - X (list of lists): The 2D feature matrix.
            - y (list of ints): The 1D binary target vector.
    """
    X = [
        [record['Transaction Amount'], record['Merchant Category Code (MCC)'], record['Transaction Response Code']]
        for record in records_list
    ]

    y = [
        record['Fraud Flag or Label']
        for record in records_list
    ]

    return X, y

X_new, y_new = build_xy_matrices(named_red)

print("First 5 entries of X_new (from function):")
print(X_new[:5])
print("\nLength of X_new:", len(X_new))

print("\nFirst 5 entries of y_new (from function):")
print(y_new[:5])
print("\nLength of y_new:", len(y_new))
```

# ** Why did you choose tuples (not lists) for your individual records?**




**Immutability**: Tuples are immutable, meaning their contents cannot be changed after creation.  Once a record is extracted, you know its values won't change.

---


**Read Named Fields**: namedtuple allows you to access fields by name (e.g., record.TransactionAmount) in addition to their index (e.g., record._1).

---


**Efficiency**: Tuples are generally more memory-efficient and slightly faster than lists, especially when dealing with a large number of fixed-size records.

---


**Use as Dictionary Keys**: Because tuples are immutable, they can be used as keys in dictionaries, which lists cannot be.

**From your DataCamp course: which of the four pandas row-iteration methods (iloc, iterrows, itertuples, apply) would you guess is fastest for processing all your records — and why might that connect to what you built in Task 1-3 today?**


---



**itertuples()**: This method returns an iterator that yields namedtuple objects (or regular tuples if index=False and name is None) for each row. namedtuple objects are much more lightweight and efficient than pandas Series objects, which iterrows yields. This significantly reduces overhead, making itertuples much faster.
iloc: iloc is primarily for integer-location based indexing to select subsets of data, not for iterating over rows efficiently. While you can loop through df.iloc[i] for each i, it's not optimized for full row-wise iteration.


---


**iterrows()**: This method yields (index, Series) pairs for each row. The constant creation of a new pandas Series object for each row is computationally expensive, making it generally the slowest method for row iteration.


---


**apply()**: The speed of apply() depends heavily on the function being applied. If the function can be vectorized (i.e., operates on entire columns at once), apply() can be very fast. However, if apply() is used with a function that processes each row individually, it often performs similarly to or worse than iterrows() because it still involves Python-level iteration and object creation overhead, though it can be slightly better due to some internal optimizations compared to explicit Python loops.
